```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from mlxtend.plotting import plot_decision_regions
import pickle
```

```python
df = pd.read_csv('contents/placement.csv')
```

```python
df.head()
```

```python
df = df.iloc[:,1:]
```

```python
df.head()
```

# Steps

# 1) Preprocess + EDA + Feature Selection
# 2) Extract input and output col
# 3) Scale the value
# 4) Train test split
# 5) Train the model
# 6) Model Selection
# 7) Deploy the model

```python
df.info()
```

```python
df.shape
```

```python
plt.scatter(df['cgpa'],df['iq'],c=df['placement'])
plt.xlabel('cgpa')
plt.ylabel('iq')
plt.title('Placement Analysis')
```

```python
input_x = df.iloc[:,:2]
output_y = df.iloc[:,-1]
```

```python
input_x
```

```python
output_y
```

```python
x_test, x_train, y_test, y_train = train_test_split(input_x, output_y, train_size=0.1)
```

```python
x_test
```

```python
x_train
```

```python
y_test
```

```python
x_test
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
x_test
```

```python
algo = LogisticRegression()
```

```python
algo.fit(x_train,y_train)
```

```python
y_pred = algo.predict(x_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
plot_decision_regions(x_train, y_train.values, clf=algo, legend=2)
```

```python
pickle.dump(algo,open('model.pkl','wb'))
```
